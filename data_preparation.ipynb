```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("student.csv")
```

```python
df
```

```python
# เปลี่ยนชื่อเพศให้เหมือนกัน
def standardize_faculty(name):
    if name in ['ชาย', 'ช', 'Male', 'M']:
        return 'ชาย'
    elif name in ['หญิง', 'ญ', 'Female', 'F']:
        return 'หญิง'
    else:
        return name

# เปลี่ยนชื่อเพศในตราง
df['เพศ'] = df['เพศ'].apply(standardize_faculty)

# เติมค่าแทนที่ NaN
df['เพศ'].fillna(df['เพศ'].mode()[0], inplace=True)
```

```python
# เปลี่ยนชื่อคณะให้เหมือนกัน
def standardize_faculty(name):
    if name in ['วิทยาศาสตร์', 'คณะวิทยาศาสตร์', 'วิทย์', 'Science', 'Sci']:
        return 'วิทยาศาสตร์'
    elif name in ['วิศวกรรมศาสตร์', 'วิศวกรรม', 'วิศวฯ', 'วิศวะ', 'Engineering']:
        return 'วิศวกรรมศาสตร์'
    elif name in ['ศิลปศาสตร์', 'ศิลป์', 'Liberal Arts', 'Arts']:
        return 'ศิลปศาสตร์'
    else:
        return name

# เปลี่ยนชื่อคณะในตราง
df['คณะ'] = df['คณะ'].apply(standardize_faculty)

# เติมค่าแทนที่ NaN
df['คณะ'].fillna(df['คณะ'].mode()[0], inplace=True)
```

```python
Q1 = df['อายุ'].quantile(0.25)
Q3 = df['อายุ'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df_no_outliers = df[(df['อายุ'] >= lower_bound) & (df['อายุ'] <= upper_bound) &
                    (df['อายุ'] >= 18) & (df['อายุ'] <= 23)]

print(f"\nค่าเฉลี่ยอายุ (หลังกรองค่าผิดปกติ) = {df_no_outliers['อายุ'].mean():.2f}")

# ค่าเฉลี่ยอายุ
mean_age = df_no_outliers['อายุ'].mean()

# เติมค่าแทนที่ NaN
df['อายุ'].fillna(mean_age, inplace=True)

# แทนที่ค่าช่วงที่เกินหรือต่ำกว่า 18-23
df.loc[(df['อายุ'] < 18) | (df['อายุ'] > 23), 'อายุ'] = mean_age
```

```python
print(f"- GPA ผิดปกติ - {df[(df['GPA'] < 0) | (df['GPA'] > 4)]['GPA'].tolist()}")

# ค่าเฉลี่ย GPA
mean_gpa = df['GPA'].mean()

# แทนที่ค่าช่วงที่เกินหรือต่ำกว่า 0 - 4
df.loc[df['GPA'] < 0, 'GPA'] = np.nan
df.loc[df['GPA'] > 4, 'GPA'] = 4.0

# เติมค่าแทนที่ NaN
df['GPA'].fillna(mean_gpa, inplace=True)
```

```python
max_height = df['ส่วนสูง_cm'].max()
min_height = df['ส่วนสูง_cm'].min()

print(f"ส่วนสูง_cm: สูงสุด = {max_height}, ต่ำสุด = {min_height}")

# ค่าเฉลี่ยส่วนสูง
mean_cm = df['ส่วนสูง_cm'].mean()

# เติมค่าแทนที่ NaN
df['ส่วนสูง_cm'].fillna(mean_cm, inplace=True)
```

```python
max_weight = df['น้ำหนัก_kg'].max()
min_weight = df['น้ำหนัก_kg'].min()

print(f"น้ำหนัก_kg: สูงสุด = {max_weight}, ต่ำสุด = {min_weight}")

# ค่าเฉลี่ยน้ำหนัก
mean_wt = df['น้ำหนัก_kg'].mean()

# เติมค่าแทนที่ NaN
df['น้ำหนัก_kg'].fillna(mean_wt, inplace=True)
```

```python
max_hours = df['ชั่วโมงเรียน_ต่อสัปดาห์'].max()
min_hours = df['ชั่วโมงเรียน_ต่อสัปดาห์'].min()

print(f"ชั่วโมงเรียน_ต่อสัปดาห์: สูงสุด = {max_hours}, ต่ำสุด = {min_hours}")

# # ค่าเฉลี่ยชั่วโมงเรียน
mean_hr = df['ชั่วโมงเรียน_ต่อสัปดาห์'].mean()

# เติมค่าแทนที่ NaN
df['ชั่วโมงเรียน_ต่อสัปดาห์'].fillna(mean_hr, inplace=True)
```
